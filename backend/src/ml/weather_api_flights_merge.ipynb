## Final merge between weather data, and flight data.

```python
# Imports

import pandas as pd
from pathlib import Path
import sys
import csv

DATA_DIR = Path().cwd()/"../../data"
WEATHER_API_DATA = DATA_DIR/"output/flights_airports_weather.csv"
FLIGHTS_DATA = DATA_DIR/"flight/cleaned_flight_sample_data_with_airports.csv"
```

```python
# Load in the .csv's

flight_data = pd.read_csv(FLIGHTS_DATA)
weather_api_data = pd.read_csv(WEATHER_API_DATA)
```

### Merge requires 2 seperate merges. One for all origin, one for all destination

```python
# Rename all cols for the two unique merges
weather_api_origin_data = weather_api_data.rename(columns={col:f"origin_{col}" for col in weather_api_data.columns})
weather_api_dest_data = weather_api_data.rename(columns={col:f"dest_{col}" for col in weather_api_data.columns})

# Merge origin:
flight_data = flight_data.merge(
    weather_api_origin_data,
    left_on=["fl_date", "origin_lat", "origin_long"],
    right_on=["origin_date", "origin_latitude", "origin_longitude"],
    how="left",
    validate="many_to_many"
)#.drop_duplicates().dropna().drop(columns=["origin_date", "origin_daily_time"])

print(flight_data.head())
```
