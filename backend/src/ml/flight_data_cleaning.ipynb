## Exploring data:

```python
# Store the sample in var
sf_df = pd.read_csv(SAMPLE_DATA_PATH)
airports = pd.read_csv(AIRPORTS_PATH)

sf_df.columns
```

### Things needed done to clean the data

- Cols to keep for *sf_df* DataFrame: ["fl_date" "op_carrier_fl_num", "op_unique_carrier", "op_carrier_fl_num", "origin", "dest", "crs_dep_time", "dep_time", "dep_delay", "taxi_out", "taxi_in", "wheels_off", "wheels_on", "arr_delay", "cancelled", "carrier_dealy" "weather_delay", "nas_delay", "security_delay", "late_aircraft_delay", "crs_arr_time"]

- Colls to keep in *airports* DataFrame: ["latitude_deg", "longitude_deg", "iata_code"]

```python
# Chop the colls:
sf_df_c = sf_df[
    [
        "fl_date",
        "op_carrier_fl_num",
        "op_unique_carrier",
        "op_carrier_fl_num",
        "origin",
        "dest",
        "crs_dep_time",
        "crs_arr_time",
        "dep_time",
        "dep_delay",
        "taxi_out",
        "taxi_in",
        "wheels_off",
        "wheels_on",
        "arr_delay",
        "cancelled",
        "carrier_delay",
        "weather_delay",
        "nas_delay",
        "security_delay",
        "late_aircraft_delay",
    ]
]
```

### These need merging, so I can run the data through the weather api.

Here, I take the airports DF which has the coords needed.

Then, make a copy with cols renamed.

Then merge this copy with the larger DF, matching it by airport code.

This is then repeated for the destination.

```python
# First merge for origin airports

rename_origin = airports.rename(columns={
    "latitude_deg":"origin_lat",
    "longitude_deg":"origin_long"
})
df = sf_df_c.merge(
    rename_origin[
        ["origin_lat", "origin_long", "iata_code"] # the cols to bring over
    ],
    left_on="origin", # the col of the origional to match to
    right_on="iata_code" # the col in the newly copied db with the code i want to match
).drop(columns=["iata_code"])

rename_dest = airports.rename(columns={
    "latitude_deg":"dest_lat",
    "longitude_deg":"dest_long"
})
df = df.merge(
    rename_dest[
        ["dest_long", "dest_lat", "iata_code"]
    ],
    left_on="dest",
    right_on="iata_code"
).drop(columns=["iata_code"])
```

### Then save the data to a new data folder in this project

```python
try:
    df.to_csv('../../data/cleaned_data.csv', index=False)
except:
    "error saving to '../../data/cleaned_data.csv'"
finally:
    print("Process has terminated")
```
